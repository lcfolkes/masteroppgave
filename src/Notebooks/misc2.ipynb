```python
def scale_up_distribution(dist: [float], percentage: float):
    for i in range(len(dist)):
        frac = dist[i] * percentage
        dist[i] -= frac
        if i == len(dist) - 1:
            dist.append(frac)
        else:
            dist[i + 1] += frac
    return dist
```
