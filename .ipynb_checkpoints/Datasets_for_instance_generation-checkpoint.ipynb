## Preparing a list to be filled with distributions. It has one row per combination of zone, weekday and hour

```python
list = []
for i in range(254):
    for j in range (3):
        list.append([i+1])
        
switch = 0
for sub_list in list:
    if (switch == 0):
        sub_list.append("Morning")
        switch_new = 1
    elif (switch == 1):
        sub_list.append("Day")
        switch_new = 2
    elif (switch == 2):
        sub_list.append("Evening")
        switch_new = 0
    switch = switch_new
    
list
```

```python
morning_list_start = vydata.loc[(vydata['TripStage'] == 'Start')
                                                  & (pd.to_datetime(vydata['DateTime']).dt.hour >= 6)
                                                  & (pd.to_datetime(vydata['DateTime']).dt.hour <= 9)]
day_list_start = vydata.loc[(vydata['TripStage'] == 'Start')
                                                  & (pd.to_datetime(vydata['DateTime']).dt.hour >= 10)
                                                  & (pd.to_datetime(vydata['DateTime']).dt.hour <= 14)]
evening_list_start = vydata.loc[(vydata['TripStage'] == 'Start')
                                                  & (pd.to_datetime(vydata['DateTime']).dt.hour >= 15)
                                                  & (pd.to_datetime(vydata['DateTime']).dt.hour <= 20)]
morning_list_end = vydata.loc[(vydata['TripStage'] == 'End')
                                                  & (pd.to_datetime(vydata['DateTime']).dt.hour >= 6)
                                                  & (pd.to_datetime(vydata['DateTime']).dt.hour <= 9)]
day_list_end = vydata.loc[(vydata['TripStage'] == 'End')
                                                  &(pd.to_datetime(vydata['DateTime']).dt.hour >= 10)
                                                  & (pd.to_datetime(vydata['DateTime']).dt.hour <= 14)]
evening_list_end = vydata.loc[(vydata['TripStage'] == 'End')
                                                  &(pd.to_datetime(vydata['DateTime']).dt.hour >= 15)
                                                  & (pd.to_datetime(vydata['DateTime']).dt.hour <= 20)]

count = 1

for line in list:
    if (count % 10 == 0):
        print(count)
    distribution_pickups = []
    distribution_deliveries = []
    
    if (line[1] == 'Morning'): 
        morning_list_start_zone = morning_list_start.loc[(morning_list_start['Zone'] == line[0])]
        morning_list_end_zone = morning_list_end.loc[(morning_list_end['Zone'] == line[0])]
        for day in range(1, 366):
            distribution_pickups.append(len(morning_list_start_zone.loc[
                                                    (pd.to_datetime(morning_list_start_zone['DateTime']).dt.dayofyear == day)]))
            
            distribution_deliveries.append(len(morning_list_end_zone.loc[
                                                    (pd.to_datetime(morning_list_end_zone['DateTime']).dt.dayofyear == day)]))
    
    elif (line[1] == 'Day'):
        day_list_start_zone = day_list_start.loc[(day_list_start['Zone'] == line[0])]
        day_list_end_zone = day_list_end.loc[(day_list_end['Zone'] == line[0])]
        for day in range(1, 366):
            distribution_pickups.append(len(day_list_start_zone.loc[
                                                    (pd.to_datetime(day_list_start_zone['DateTime']).dt.dayofyear == day)]))
            
            distribution_deliveries.append(len(day_list_end_zone.loc[
                                                    (pd.to_datetime(day_list_end_zone['DateTime']).dt.dayofyear == day)]))
            
    elif (line[1] == 'Evening'):
        evening_list_start_zone = evening_list_start.loc[(evening_list_start['Zone'] == line[0])]
        evening_list_end_zone = evening_list_end.loc[(evening_list_end['Zone'] == line[0])]
        for day in range(1, 366):
            distribution_pickups.append(len(evening_list_start_zone.loc[
                                                    (pd.to_datetime(evening_list_start_zone['DateTime']).dt.dayofyear == day)]))
            
            distribution_deliveries.append(len(evening_list_end_zone.loc[
                                                    (pd.to_datetime(evening_list_end_zone['DateTime']).dt.dayofyear == day)]))
            
    
    distribution_pickups.sort()
    distribution_deliveries.sort()
    counter_pickups = {item:distribution_pickups.count(item) for item in distribution_pickups}
    
    counter_deliveries = {item:distribution_deliveries.count(item) for item in distribution_deliveries}  
    counter_pickups = json.dumps(counter_pickups)
    counter_deliveries = json.dumps(counter_deliveries)
    line.append(counter_pickups)
    line.append(counter_deliveries)
    
    count+=1
    
```
